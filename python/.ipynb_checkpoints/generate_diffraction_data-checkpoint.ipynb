```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
from propagators import FresnelPropagator
```

### Experimental parameters

```python
wavelength = 1.3776e-10 # meters == 9 keV
distance = 15 # meters
pixels_size = 55e-6 # meters
```

### Create object and probe

```python
magnitude = getattr(data, 'camera')()
phase = getattr(data, 'gravel')()
phase = phase/np.max(phase)
obj_model = np.abs(magnitude)*np.exp(1j*phase)
```

```python
probe = np.zeros_like(obj_model)
probe_radius = 40
y_coor = np.linspace(0,probe.shape[0]-1,probe.shape[0]) - probe.shape[0]//2
x_coor = np.linspace(0,probe.shape[1]-1,probe.shape[1])  - probe.shape[1]//2
Ymesh, Xmesh = np.meshgrid(y_coor,x_coor)
probe = np.where(Xmesh**2 + Ymesh**2 <= probe_radius**2, 1, 0)

object_pixel_size = 1e-6
probe_phase = FresnelPropagator(probe, object_pixel_size, wavelength, 100e-2)
probe = np.abs(probe_phase)*np.exp(1j*probe_phase)
```

```python
fig, ax = plt.subplots(2,2,dpi=100)
ax0 = ax[0,0].imshow(np.abs(obj_model))
ax1 = ax[0,1].imshow(np.angle(obj_model))
ax2 = ax[1,0].imshow(np.abs(probe))
ax3 = ax[1,1].imshow(np.angle(probe))
```
